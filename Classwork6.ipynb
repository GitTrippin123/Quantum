```python
#preamble
import sympy as sym
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
from sympy.plotting import plot
sym.init_printing()
from sympy import Symbol, oo, pi, sin, cos, lambdify, latex,\
                    Derivative, diff, Integral, integrate,\
                    solve, expand, sqrt, simplify
from IPython.display import display, Math, Latex
plt.style.use(['dark_background'])

x=Symbol("x", real=True)

h=Symbol("hbar",real=True, positive=True, nonzero = True)
m=Symbol("m", real=True, positive=True, nonzero = True)
w=Symbol("omega", real=True, positive=True, nonzero = True)

xi = (m*w/h)**(1/2)*x

q=Symbol("q", real=True)
s=Symbol("s", real=True)
```

```python
def herm(n): 
    p = sci.special.hermite(n,monic=False)
    return sum(coef * x**i for i, coef in enumerate(p.coeffs[::-1]))
    
    #returns nth Hermite polynomial as numpy.polynomial

def psiSHO(v):
    
    N = sym.sqrt(1/(2**(v)*sym.factorial(v)))*(m*w/(h*sym.pi))**(1/4)
    herm_poly = herm(v).subs(x, xi)
    
    Psi = N*herm_poly*sym.exp(-xi**2/2)
    #display(Math(latex(Psi)))
    return Psi
    
    #returns normalized SHO wavefunction with quantum number v

def energySHO(v):
    return h*w*(v+1/2)

#verification
Wavefunctions = [psiSHO(i) for i in range(8)]
Energys = [energySHO(i) for i in range(8)]
#display(Math(latex(Wavefunctions[2])))

#computing the second order perturbation summation
terms = []
for i in range(8):
    if i==3:
        continue
    else:
        numerator = (sym.integrate(psiSHO(3)*(q*x+s*x**4)*psiSHO(i),(x,-sym.oo,sym.oo)))**2
        denominator = energySHO(3)-energySHO(i)
        terms.append(numerator/denominator)
    #display(Math(latex(numerator)))

E2_3 = sym.simplify(sum(terms))

#for i in terms:
#    display(Math(latex(sym.simplify(i))))

display(Math(latex(E2_3)))

```

```python
rho = Symbol("rho", real=True)
Z = Symbol("Z", real=True)
phi = Symbol("phi", real=True)
theta = Symbol("theta", real=True)

Li321 = - 3**(7/2)*rho**2 /(162*sym.sqrt(sym.pi))*sym.exp(-3*rho/3)*sym.sin(2*theta)
wsq = Li321**2

display(Math(latex(wsq)))
```

```python
etheta = sym.integrate(wsq*rho**2*sym.sin(theta)*theta**2,(theta,0,sym.pi))

he = sym.integrate(etheta,(rho,0,sym.oo))

h = sym.integrate(he,(phi,0,2*sym.pi))

display(Math(latex(sym.simplify(h))))
```
