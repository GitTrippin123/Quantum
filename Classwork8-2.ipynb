### define some symbols

```python
r=sym.Symbol("r", real=True, positive=True)
rho=sym.Symbol("rho", real=True, positive=True)

R=sym.Symbol("R", real=True, positive=True)
z=sym.Symbol("zeta", positive=True, nonzero=True)
```

### Define Integrals symbolically

```python
s = (1 + R + (R**2)/3)*sym.exp(-R)
s
```

```python
taa = Rational(1,2)*z**2
taa
```

```python
vaaa = -z
vaaa
```

```python
tab = z**2*(1 + R - (R**2)/3)*sym.exp(-R)/2
tab
```

```python
vaba =  -z*(1+R)*sym.exp(-R)
vaba
```

```python
vaab = -z*(1-(1+R)*sym.exp(-2*R))/R
vaab
```

```python
E = 1/(1+s)*(taa+tab+vaaa+vaab+2*vaba) + z/R

Ed = sym.diff(E,z)
sol = sym.solve(Ed,z)[0]

eq = sol.subs(R,z*rho)

plt.rcParams.update(plt.rcParamsDefault)

plot(sol,(R,0.1,7), axis_center=(0,0), ylim = (-.1,0.2),xlabel="R",ylabel="E")
```
